# The Basic Structure of a Seaborn Plot

## Make Your First Plot

```python
data  = sb.load_dataset('titanic')
data.head()
```

```python
sb.countplot(data=data,y='sex',hue='alive')
```

## Lets Make Plots Focusing on 1 Variable (Univariate Analysis )

**Univariate Plots**
1.   Continuous 
  *   Histogram
  *   Boxplot
  *   Boxenplot
  *   Rugplot
  *   Violinplot
  *   KDE Plot
  *   Strip Plot

2.   Categorical
  *   Countplot
  







```python
data = sb.load_dataset('exercise')
data.head()
```

```python
sb.histplot(data =data,x='pulse')
```

```python
sb.boxplot(data =data,x='pulse')
```

```python
sb.boxenplot(data =data,x='pulse')
```

```python
sb.violinplot(data =data,x='pulse')
```

```python
sb.rugplot(data =data,x='pulse')
```

```python
sb.kdeplot(data =data,x='pulse',shade=True)
```

```python
sb.stripplot(data =data,x='pulse')
```

```python
sb.swarmplot(data =data,x='pulse')
```

```python
sb.ecdfplot(data =data,x='pulse')
```

## Using the Grouping Variable (HUE)

```python
sb.ecdfplot(data =data,x='pulse',hue='diet')
```

## Combining to Univariate Charts

```python
sb.kdeplot(data =data, x = 'pulse',hue = 'diet')
sb.rugplot(data =data, x = 'pulse',hue = 'diet')
```

## Lets Make Plots Focusing on 2 Variables (Bivariate Analysis )

**Bivariate Plots**
1.   Continuous 
  *   Scatterplot
  *   Lmp Plot
  *   Regplot
  *   Hex Plot
  *   Pair Grid
  *   KDE Plot
  
2.   Categorical
  *   Line Plot
  *   Bar Plot
  *   Point Plot
  







```python
data = sb.load_dataset('flights')
data.head()
```

```python
sb.lineplot(data=data,x='year',y='passengers')
```

```python
sb.pointplot(data=data,x='year',y='passengers')
```

```python
sb.barplot(data=data,x='year',y='passengers')
```

```python
data = sb.load_dataset('diamonds')
data.head()
```

```python
data = data.sample(3000)
```

```python
sb.scatterplot(data=data,x='carat',y='price')
```

```python
sb.scatterplot(data=data,x='carat',y='price', marker="+")
```

```python
sb.kdeplot(data=data,x='carat',y='price')
```

```python
sb.kdeplot(data=data,x='carat',y='price',fill=True,cmap='mako_r')
```

```python
sb.histplot(data=data,x='carat',y='price')
```

```python
sb.kdeplot(data=data,x='carat',y='price',cmap='coolwarm')
sb.histplot(data=data,x='carat',y='price',color='red')
```

```python
sb.jointplot(data=data,x='carat',y='price')
```

```python
sb.jointplot(data=data,x='carat',y='table',kind='hex')
```

```python
sb.lmplot(data=data,x='carat',y='price',markers=".",line_kws={'color': 'red'})
```

## Plotting Relationships

```python
sb.relplot(data=data,x='carat',y='price',col='cut')
```

```python
data = sb.load_dataset('tips')
data.head()
```

```python
facet = sb.FacetGrid(data=data, col='smoker',row='sex')
facet.map(sb.boxplot,'total_bill')
```

```python
sb.pairplot(data)
```

## Correlations

```python
data=sb.load_dataset('diamonds')
data.head()
```

```python
sb.heatmap(data.corr())
```

```python
sb.heatmap(data.corr(),annot=True,linewidths=1)
```

```python
sb.heatmap(data.corr()[['price']],annot=True,linewidths=1)
```
